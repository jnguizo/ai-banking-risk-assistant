# EDA — AI Banking Risk Assistant
**Dataset : Credit Card Fraud Detection**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
df = pd.read_csv('../data/creditcard.csv')
print('Shape:', df.shape)
df.head()
```

## 1. Aperçu général

```python
print('--- Infos ---')
print(df.info())
print('\n--- Valeurs manquantes ---')
print(df.isnull().sum().sum())
print('\n--- Statistiques ---')
df.describe()
```

## 2. Distribution des fraudes

```python
fraud = df['Class'].value_counts()
print(fraud)
print(f'\nTaux de fraude : {fraud[1]/len(df)*100:.2f}%')

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Class', palette='Set2')
plt.xticks([0,1], ['Normal','Fraude'])
plt.title('Distribution des transactions')
plt.savefig('../reports/distribution_fraude.png')
plt.show()
```

## 3. Distribution des montants

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
sns.histplot(df[df['Class']==0]['Amount'], bins=50, color='blue', alpha=0.6)
plt.title('Montants - Transactions normales')
plt.xlabel('Montant')

plt.subplot(1,2,2)
sns.histplot(df[df['Class']==1]['Amount'], bins=50, color='red', alpha=0.6)
plt.title('Montants - Transactions frauduleuses')
plt.xlabel('Montant')

plt.tight_layout()
plt.savefig('../reports/distribution_montants.png')
plt.show()
```

## 4. Corrélation avec la fraude

```python
corr = df.corr()['Class'].drop('Class').sort_values()

plt.figure(figsize=(10,8))
corr.plot(kind='barh', color=['red' if x < 0 else 'blue' for x in corr])
plt.title('Corrélation des variables avec la fraude')
plt.tight_layout()
plt.savefig('../reports/correlation_fraude.png')
plt.show()
```
